```
rdd=sc.parallelize([(1,'a'),(2,'c'),(1,'b')])
print("Original RDD :", rdd.collect())
print("After transformation : ", rdd.sortByKey().collect()) # ReduceByKey : add the elements of same key values
```

```
rdd=sc.parallelize([(1,'a'),(2,'c'),(1,'b')])
print("Original RDD :", rdd.collect())
print("After transformation : ", rdd.groupByKey().collect())
```

```
rdd = sc.parallelize([(1,2), (2,4), (2,6)])
print("Original RDD :", rdd.collect())
print("After transformation : ", rdd.mapValues(lambda x: x*x).collect())
```

```
rdd = sc.parallelize([(1,2), (2,4), (2,6)])
print("Original RDD :", rdd.collect())
#print("After transformation : ", rdd.groupByKey().mapValues(lambda x:[a for a in x]).collect())
print("After transformation : ", rdd.groupByKey().mapValues(lambda x:[a+a for a in x]).collect()) # narrow + wide transformation
print("After transformation groupByKey: ", rdd.groupByKey().mapValues(sum).collect()) # to add values
print("After transformation reduceByKey: ", rdd.reduceByKey(lambda x,y: x+y).collect()) # reduce by key
# reduce by key - shuffle less data, intermediate shuffle happen inside the partition, only the partitions need shuffle will be shuffled
# group by key - always shuffle the data
# aggregate by key - to change the context of values
```

```
rdd = sc.parallelize([(1,2), (2,4), (2,6)])
print("Original RDD :", rdd.collect())
# the lambda function generates for each number i, an iterator that produces i,i+1
print("After transformation : ", rdd.flatMapValues(lambda x: list(range(x,x+2))).collect())
```
